#### The map() function
Set the values of a column according to the values in another column. Depending on the value of the 'current_location' column, we set the value in a new colun called 'import'. A dictionary is used for this

```python
cars_data['current_location'].unique()
```

```python
cars_data['import'] = cars_data['current_location']\
                            .map({'In Tanzania' : 'No', 
                                  'Available For Import': 'Yes'})
```

```python
cars_data.head(5)
```

```python
cars_data['import'].unique()
```

#### Defining an external function to remove commas

```python
def drop_comma(x):
    return str(x).replace(',','')
```

```python
cars_data['price_value'].head(10)
```

```python
cars_data['price_value'] = cars_data['price_value'].apply(drop_comma)
```

#### The prices don't have commas any more
But they're still of type 'object'

```python
cars_data['price_value'].head(10)
```

```python
cars_data['price_value'] = pd.to_numeric(cars_data['price_value'])
```

```python
cars_data['price_value'].head(10)
```

Defining an external function that categorises the cars according to price

```python
def category(price):
    
    price = int(price)
    
    if price <= 10000000:
        return 'cheap'
    elif price > 10000000 and price <= 30000000:
        return 'mid-range'
    elif price > 30000000 and price <= 70000000:
        return 'expensive'
    else:
        return 'luxury'
```

#### Apply function to numeric column
Fields of a dataframe can also be referenced using the dot operator

```python
cars_data.price_value.apply(category).head(10)
```

Adding this as a column to our dataframe

```python
cars_data['price_range'] = cars_data.price_value.apply(category)
```

```python
cars_data.head(10)
```

Which is the costliest car in the dataframe? Which car has the most mileage?

```python
cars_data[['price_value', 'mileage']].apply(max)
```

```python
cars_data.iloc[cars_data['price_value'].idxmax()].values
```

```python
cars_data.iloc[cars_data['mileage'].idxmax()].values
```

What we have done is applied the max function for all columns from price_value to mileage. Thus we have max values for every column in this list

#### Applying the window function

```python
np_data.rolling(window = 3).mean()
```

window size = 3; applying the mean for this window. COnsiders 3 elements at a time. Gives average of the n, n-1 and n-2 elements

#### Applying the expanding function

```python
np_data.expanding(min_periods = 3).mean()
```

The first average is for 3 elements. next is for 4, next 5 ... is expanded to included further elements (from further rows) with each computation
